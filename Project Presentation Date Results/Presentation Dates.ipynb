## Results of Project Presentation Dates

```python
import pandas as pd
import random
xls = pd.ExcelFile("Teams.xlsx")
teams = xls.parse('Sheet1', index_col=None, na_values=['NA'])
teams
```

```python
indices = list(teams.index)
indices
```

```python
random.shuffle(indices)
indices
```

```python
randomShuffled = pd.DataFrame(data=teams, index=indices)
randomShuffled
```

```python
newOrder = randomShuffled.reset_index(drop=True, inplace=False)
newOrder
```

```python
dates = pd.Series(range(0,20))
dates[:10] = '03/12/2015'
dates[10:] = '10/12/2015'
dates.name = 'Presentation Date'
newOrder.join(dates)
```
